```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import os
import random
from tensorflow.keras import layers, models
from PIL import Image
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = image / 255.0
    return image

def load_dataset(root_directory, dataset_type):
    dataset_directory = os.path.join(root_directory, 'data', dataset_type)
    images = []
    for folder_name in os.listdir(dataset_directory):
        folder_path = os.path.join(dataset_directory, folder_name)
        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                if filename.endswith('.jpg'):
                    image_path = os.path.join(folder_path, filename)
                    image = preprocess_image(image_path)
                    images.append(image)
    return images

root_directory = 'moon'
train_images = load_dataset(root_directory, 'train')
test_images = load_dataset(root_directory, 'test')
val_images = load_dataset(root_directory, 'val')

print("Number of training images: ", len(train_images))
print("Number of tet images: ", len(test_images))
print("Number of validation images:", len(val_images))
```

```python
def display_random_images(images, num_images = 5):
    fig, axes = plt.subplots(1, num_images, figsize = (15, 5))
    for i in range(num_images):
        random_index = random.randint(0, len(images) - 1)
        image = images[random_index]
        image_unit8 = (image * 255).astype('uint8')
        axes[i].imshow(image_unit8)
        axes[i].axis('off')
    plt.show()

display_random_images(train_images, num_images = 5)
display_random_images(test_images, num_images = 5)
display_random_images(val_images, num_images = 5)
```

# Lunar Dreams: Generating Realistic Moon Images using Deep Learning 

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import cv2
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = image / 255.0
    return image

def load_dataset(root_directory, dataset_type):
    dataset_directory = os.path.join(root_directory, 'data', dataset_type)
    images = []
    for folder_name in os.listdir(dataset_directory):
        folder_path = os.path.join(dataset_directory, folder_name)
        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                if filename.endswith('.jpg'):
                    image_path = os.path.join(folder_path, filename)
                    image = preprocess_image(image_path)
                    images.append(image)
    return images

root_directory = 'moon'
train_images = load_dataset(root_directory, 'train')
test_images = load_dataset(root_directory, 'test')
val_images = load_dataset(root_directory, 'val')

print("Number of training images:", len(train_images))
print("Number of test images:", len(test_images))
print("Number of validation images:", len(val_images))
```

```python
def display_random_images(images, num_images = 5):
    fig, axes = plt.subplots(1, num_images, figsize = (15, 5))
    for i in range(num_images):
        random_index = random.randint(0, len(images) - 1)
        image = images[random_index]
        image_uint8 = (image * 255).astype('uint8')
        axes[i].imshow(image_uint8)
        axes[i].axis('off')
    plt.show()

display_random_images(train_images, num_images = 5)
display_random_images(test_images, num_images = 5)
display_random_images(val_images, num_images = 5)
```

```python
tf.random.set_seed(42)
np.random.seed(42)

def build_generator(input_shape = (100,), output_channels = 3):
    model = models.Sequential([
        layers.Dense(8 * 8 * 256, input_shape = input_shape),
        layers.Reshape((8, 8, 256)),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Conv2DTranspose(128, kernel_size = 5, strides = 2, padding = 'same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Conv2DTranspose(64, kernel_size = 5, strides = 2, padding = 'same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Conv2DTranspose(output_channels, kernel_size = 5, strides = 2, padding = 'same', activation = 'tanh')
    ])
    return model

def build_discriminator(input_shape = (64, 64, 3)):
    model = models.Sequential([
        layers.Conv2D(64, kernel_size = 5, strides = 2, padding = 'same', input_shape = input_shape),
        layers.LeakyReLU(),
        layers.Dropout(0.3),

        layers.Conv2D(128, kernel_size = 5, strides = 2, padding = 'same'),
        layers.LeakyReLU(),
        layers.Dropout(0.3),

        layers.Conv2D(256, kernel_size = 5, strides = 2, padding = 'same'),
        layers.LeakyReLU(),
        layers.Dropout(0.3),

        layers.Flatten(),
        layers.Dense(1, activation = 'sigmoid')
    ])
    return model

def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = models.Sequential([generator, discriminator])
    return model

def preprocess_images(images):
    images = np.array(images)
    images = (images.astype('float32') - 127.5) / 127.5
    return images

def load_dataset(root_directory, dataset_type):
    dataset_directory = os.path.join(root_directory, 'data', dataset_type)
    images = []
    for folder_name in os.listdir(dataset_directory):
        folder_path = os.path.join(dataset_directory, folder_name)
        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                if filename.endswith('.jpg'):
                    image_path = os.path.join(folder_path, filename)
                    image = tf.io.read_file(image_path)
                    image = tf.image.decode_jpeg(image, channels = 3)
                    image = tf.image.resize(image, (64, 64))
                    images.append(image)
    return preprocess_images(images)

def display_generated_images(generator, num_images = 5, input_shape = (100,), output_channels = 3):
    noise = tf.random.normal([num_images, *input_shape])
    generated_images = generator.predict(noise)
    plt.figure(figsize = (10, 10))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(generated_images[i] * 0.6 + 0.6)
        plt.axis('off')
    plt.show()

batch_size = 128
epochs = 120
input_shape = (100,)
root_directory = 'moon'

train_images = load_dataset(root_directory, 'train')

discriminator = build_discriminator()
discriminator.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0002, beta_1 = 0.5),
                      loss='binary_crossentropy', metrics = ['accuracy'])

generator = build_generator()

gan = build_gan(generator, discriminator)
gan.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0002, beta_1 = 0.5),
            loss = 'binary_crossentropy')

for epoch in range(epochs):
    for batch in range(len(train_images) // batch_size):
        noise = tf.random.normal([batch_size, *input_shape])
        fake_images = generator.predict(noise)
        real_images = train_images[batch * batch_size: (batch + 1) * batch_size]

        discriminator_loss_real = discriminator.train_on_batch(real_images, np.ones((batch_size, 1)))
        discriminator_loss_fake = discriminator.train_on_batch(fake_images, np.zeros((batch_size, 1)))
        discriminator_loss = 0.5 * np.add(discriminator_loss_real, discriminator_loss_fake)

        noise = tf.random.normal([batch_size, *input_shape])
        gan_loss = gan.train_on_batch(noise, np.ones((batch_size, 1)))

    print(f'Epoch {epoch + 1}, Discriminator Loss: {discriminator_loss[0]}, Generator Loss: {gan_loss}')

    if (epoch + 1) % 10 == 0:
        display_generated_images(generator)  #TODO: Fix till realistic images
```

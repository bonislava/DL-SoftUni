```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import os
import random
from tensorflow.keras import layers, models
from PIL import Image
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = image / 255.0
    return image

def load_dataset(root, dtype):
    directory = os.path.join(root, 'data', dtype)
    images = []
    for moon in os.listdir(directory):
        path = os.path.join(directory, moon)
        if os.path.isdir(path):
            for filename in os.listdir(path):
                if filename.endswith('.jpg'):
                    image_path = os.path.join(path, filename)
                    image = preprocess_image(image_path)
                    images.append(image)
    return images

root_directory = 'C:\\Users\\boni_\\Desktop\\moon'
train = load_dataset(root_directory, 'train')
test = load_dataset(root_directory, 'test')
val = load_dataset(root_directory, 'val')

print("Number of training images: ", len(train))
print("Number of tet images: ", len(test))
print("Number of validation images:", len(val))
```

```python
def random_im (images, num = 5):
    fig, axes = plt.subplots(1, num, figsize = (15,5))
    for i in range(num):
        index = random.randint(0,len(images)- 1)
        image = images[index]
        image_unit8 = (image * 255).astype('uint8')
        axes[i].imshow(image_unit8)
        axes[i].axis('off')
    plt.show()

random_im(train, num = 5)
random_im(test, num = 5)
random_im(val, num = 5)
```
